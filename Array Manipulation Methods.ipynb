1. Copy in NumPy
What is Copy?

A copy creates a completely new array.

If you change the copied array, the original array does not change.

```python
import numpy as np

a = np.array([1, 2, 3, 4])

b = a.copy()

b[0] = 100

print(a)
print(b)
```

Explanation:
b = a.copy() creates a separate array
Changing b will not affect a

Think like this:

Copy = Xerox copy /
Changing xerox paper won't affect original paper.

2. View in NumPy
What is View?

A view shares memory with the original array.

If you change one, the other also changes.

Example:

```python
a = np.array([10, 20, 30])

b = a.view()

b[0] = 999

print(a)
print(b)
```

Explanation:

view() does not create new memory.

Think like:

View = Mirror /
Change original → reflection changes.

Difference Between Copy and View
Copy	          View
New memory	      Same memory
Original          won't change	Original changes
Independent	      Connected

3. Flatten
What is flatten()?

Converts multi-dimensional array into 1D array.

It creates a copy.

Example:

```python
a = np.array([[1,2],
              [3,4]])

b = a.flatten()

print(b)
```

Important:

flatten() = copy

Changing flattened array won't affect original.

```python
b[0] = 100

print(a)
```

No change.

4. Ravel
What is ravel()?

Also converts array into 1D.

But it returns a view (mostly).

Example:

```python
a = np.array([[1,2],
              [3,4]])

b = a.ravel()

b[0] = 100

print(a)
```

Difference:
flatten() → copy
ravel() → view

flatten vs ravel
flatten()	          ravel()
Copy	              View
More memory	          Less memory
Original unchanged	  Original may change

5. Reshape
What is reshape()?

Changes shape of array.

Example:

1D → 2D

```python
a = np.array([1,2,3,4,5,6])

b = a.reshape(2,3)

print(b)
```

Explanation:

(2,3) means:

2 rows
3 columns

Rule:

Total elements must match.

Correct: 6 elements = 2 × 3
Wrong: a.reshape(4,4)
Error because: 6 ≠ 16

Using -1

NumPy automatically calculates.

```python
a.reshape(2,-1)
```

6. Transpose
What is transpose?

Converts:

Rows → Columns
Columns → Rows

Example:

```python
a = np.array([[1,2,3],
              [4,5,6]])

a.T
```

Before: 2 × 3

After transpose: 3 × 2

Shortcut: a.T or a.transpose()

7. Broadcasting
What is Broadcasting?

NumPy automatically stretches smaller arrays to match bigger arrays.

Example 1:

```python
a = np.array([1,2,3])

a + 10
```

Here NumPy treats: 10 → [10,10,10]

Example 2:

```python
a = np.array([[1,2,3],
              [4,5,6]])

b = np.array([10,20,30])

a + b
```

NumPy stretches b:

[10,20,30]

to

[[10,20,30],
 [10,20,30]]

This is broadcasting.

9. Variance
What is Variance?

Measures how spread out values are from mean.

Formula idea:

Difference from mean → square → average.

Example:

```python
a = np.array([10,20,30,40])

np.var(a)
```

Meaning:

Higher variance = data spread more.

Lower variance = data close together.

11. Unique
What is unique()?

Removes duplicates.

Example:

```python
a = np.array([1,2,2,3,3,4,5])
np.unique(a)
```

12. Percentile
What is Percentile?

Tells where a value lies in data.

Example:

50th percentile = median
25th percentile = lower quarter
75th percentile = upper quarter
Example:

```python
a = np.array([10,20,30,40,50])
np.percentile(a,50)
```

```python
np.percentile(a,25)
```

13. Sum
What is sum()?

Adds all values.

Example:
