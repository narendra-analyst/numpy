1. Identity Matrix

An identity matrix is a matrix where:

diagonal = 1
remaining places = 0

Example (3×3):
1 0 0
0 1 0
0 0 1

```python
import numpy as np

i = np.identity(3)
i
```

You can also use: np.eye(3)
Same output.

Why called identity?

Because multiplying any matrix with identity gives the same matrix.

Example:

```python
A = np.array([[1,2],
              [3,4]])

I = np.identity(2)

A@I
```

A × I = A

2. A Inverse Matrix (A⁻¹)

Inverse means opposite matrix.

Formula idea:

If A × A⁻¹ = I

then A⁻¹ is inverse of A.

Example matrix:

```python
A = np.array([[1,2],
              [3,4]])
np.linalg.inv(A)
```

```python
A = np.array([[1,2],
              [3,4]])
A @ np.linalg.inv(A)
```

That is identity matrix.

So: A × A⁻¹ = I

3. I Inverse (I⁻¹)

If you mean identity matrix inverse, then:

The inverse of identity matrix is identity itself.

Because: I × I = I

```python
I = np.identity(3)

np.linalg.inv(I)
```

Same matrix.

So: I⁻¹ = I

4. Determinant Matrix

Determinant gives a single number from a square matrix.

Formula idea for 2×2 matrix:

For
A=[ a b ]
  [ c d ]
Determinant is:
∣A∣=ad−bc
Example:

```python
A = np.array([[1,2],
              [3,4]])

np.linalg.det(A)
```

Calculation:

(1×4) - (2×3)
= 4 - 6
= -2

Important Rule

Inverse exists only if determinant ≠ 0

Example:

```python
A = np.array([[1,2],
              [2,4]])
np.linalg.det(A)
```

```python
Now inverse:
A = np.array([[1,2],
              [2,4]])
np.linalg.inv(A)
```
