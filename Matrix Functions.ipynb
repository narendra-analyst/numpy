1. Matrix (2D Array) in NumPy

Before learning matrix functions, understand what a matrix is.

A matrix is simply a 2D array (rows and columns).

Example:

```python
import numpy as np

a = np.array([[1, 2],
              [3, 4]])

a
```

here:

Row = horizontal
Column = vertical

This is a 2×2 matrix

Because:

2 rows
2 columns

2. Identity Matrix

An identity matrix is a square matrix where:

Diagonal values = 1
Other values = 0

Example:

[[1 0]
 [0 1]]

Think of it like number 1 in multiplication.

Example:

```python
b = np.identity(3)
b
```

Explanation:

3 means create a 3×3 identity matrix

Another way:

np.eye(3)

Same output.

3. Matrix Transpose

Transpose means:

➡️ Rows become columns
➡️ Columns become rows

Example matrix:

[[1, 2],
 [3, 4]]

After transpose:

[[1, 3],
 [2, 4]]

```python
a = np.array([[1,2],
              [3,4]])

a.T
```

Easy Trick:

Transpose = flip across diagonal

6. Matrix Multiplication

This is important.

There are 2 types:

Type 1: Normal Multiplication

Element-by-element multiplication.

```python
a*b
```

Because:

1×5 = 5
2×6 = 12
3×7 = 21
4×8 = 32

Type 2: Matrix Multiplication (Real Matrix Math)

Use: np.dot() or @

Example:

```python
a = np.array([[1,2],
              [3,4]])

b = np.array([[5,6],
              [7,8]])

np.dot(a,b)
```

How?

First row × first column:

(1×5) + (2×7)
= 5 + 14
= 19

First row × second column:

(1×6) + (2×8)
= 6 + 16
= 22

Like that all values are calculated.

Shortcut:

print(a @ b)

Same output.

7. Determinant of Matrix

Determinant tells whether matrix can have inverse or not.

Function:

np.linalg.det()

Example

```python
a = np.array([[1,2],
              [3,4]])

np.linalg.det(a)
```

Important Rule:

If determinant = 0

Inverse cannot be found.

If determinant ≠ 0

Inverse possible.

8. Matrix Inverse

Inverse means reverse matrix.

Function:

np.linalg.inv()

Example:

```python
a = np.array([[1,2],
              [3,4]])
np.linalg.inv(a)
```

Easy Understanding:

Inverse is like division in matrix.

Example:

A × A⁻¹ = Identity Matrix

Means:

[[1,0],
 [0,1]]

Code:

```python
a = np.array([[1,2],
              [3,4]])

inverse = np.linalg.inv(a)

a @ inverse
```

This is identity matrix.

9. Matrix Power

Raise matrix to power.

Function:

np.linalg.matrix_power()

Example:

```python
a = np.array([[1,2],
              [3,4]])

np.linalg.matrix_power(a,2)
```

Means:

A × A

If:

3

Means:

A × A × A

10. Rank of Matrix

Rank tells:

➡️ Number of independent rows or columns.

Function:

np.linalg.matrix_rank()

Example:

```python
a = np.array([[1,2],
              [3,4]])

np.linalg.matrix_rank(a)
```
